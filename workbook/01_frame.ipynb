## What a problem statement looks like

The example is deliberately about something else (a coffee shop), so it can't be copied.

**Bad:**

> "We want to use data science on our sales data to find insights and optimize
> the business."

Why it's bad: no decision-maker, no decision, no measurable outcome, no scope. "Insights"
and "optimize" name a wish, not a problem. Nothing in it tells you which table to open
first, and no result could ever prove it finished.

**Good:**

> "The owner schedules two baristas on weekday mornings but queues still exceed five
> minutes between 8:00 and 9:30, and she suspects mobile orders are the cause. Using six
> months of till and mobile-order timestamps, quantify how much of the 8:00–9:30 queue
> time is attributable to mobile orders, so she can decide whether a third barista or a
> separate mobile pickup shelf is the cheaper fix."

Why it's good: a named decision-maker (the owner), a concrete decision (third barista vs
pickup shelf), the data that will be used (six months of timestamps), a measurable
quantity (attributable queue time), and a scope (weekday 8:00–9:30). You could tell the
day this analysis is done.

## Your problem statement

One paragraph. It should name who decides something with this analysis (a GM? you as
designer of custom monsters?), what they decide, what data the repo actually has
(`sd_monsters`, `sd_attacks`, `fe_monsters`, `crosswalk`, `sd_spells`, the combat-sim
results), and what measurable output would settle it.

---

As the GM of an ongoing Shadowdark campaign who also designs custom monsters and converts creatures from 5e-era material, I have to decide at prep time whether to trust a monster's printed LV or adjust it, and when writing my own stat blocks I need to know the house style I am imitating. This dataset is designed data, not organic data: every pattern in it is a decision the game's designer made, so the analysis is design forensics rather than discovery of natural behavior. Using the sd_monsters, sd_attacks, and crosswalk tables plus the combat-sim win rates in reports/, quantify how well printed LV tracks measured difficulty, identify where and for which monster types it fails, and describe the statistical house style of the core bestiary well enough that I can stat a custom monster that fits it. The analysis is done when I have a short list of concrete adjustment rules I would actually apply at the prep table.

## Your three questions

Each question must be specific enough that you could name the SQL query or plot that
answers it. "Are monsters balanced?" is not answerable; "do printed LVs order monsters
the same way simulated win rates do?" is.

For each one, fill in all three parts:

### Question 1
-Question: Setting aside HP, which Shadowdark rules derive directly from LV (one hit die per level), which monster stats carry independent information about LV, and how much?
   Predicted outcome, stated up front: a model including HP will fit almost perfectly because HP is LV in costume; the honest question is what AC, attack bonus, damage, 
   and stat mods explain on their own.
- Data that would answer it: sd_monsters (level, ac, stat mods) joined to sd_attacks (attack_bonus, avg_damage, num_attacks); correlations, then a linear regression on the 
  non-HP features with coefficients and R squared.
- Cost of a wrong answer at the table: If I stat a custom monster by copying the wrong drivers, its LV label lies. The party treats it as a fair fight and it is not, in either 
  direction: an accidental TPK or a boss that folds in two rounds.

### Question 2
- Question: Does alignment correlate with difficulty skew? In other words, did the designer systematically make chaotic monsters punch above their printed LV, or is alignment 
  flavor with no statistical fingerprint?
- Data that would answer it: sd_monsters.alignment grouped against LV-model residuals and against sim win rates from reports/, compared across L/N/C; a grouped boxplot and group 
  means, with group sizes reported since some alignments may be rare.
- Cost of a wrong answer at the table: If chaotic monsters run hot and I do not know it, encounters built from demon-heavy or cult-heavy rosters will be systematically deadlier 
  than their LV budget suggests, and my Dark Sun conversions (heavily chaotic) inherit the same hidden bias.

### Question 3
- Question: What is the shape of the bestiary's LV distribution? Did the designer balance monster counts per level (a flat design), build a pyramid with many low-LV and few high-LV 
  monsters, or something lumpier, and where are the thin spots?
- Data that would answer it: sd_monsters.level alone; a histogram of counts by LV, plus counts by LV band, optionally split by monster type from traits.
-Cost of a wrong answer at the table: This one costs prep time rather than PC lives. If levels 7 to 9 are thin and I do not know it, I will discover mid-campaign that the book 
  cannot supply varied encounters for that tier and I will be improvising stat blocks under time pressure instead of designing them properly in advance, which is exactly when bad custom monsters get made.
